```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Create a confusion matrix as a DataFrame
confusion_data = {
    'Predicted: Cheated': [3, 3],
    'Predicted: Honest': [2, 12]
}
confusion_df = pd.DataFrame(confusion_data, index=['Actually Cheated', 'Actually Honest'])

# Set up the heatmap
plt.figure(figsize=(5,3))
sns.heatmap(confusion_df, annot=True, fmt='d', cmap='YlGnBu', cbar=False, linewidths=1, linecolor='black')

# Title and labels
plt.title('Confusion Matrix Example (Detecting Cheaters)', fontsize=14)
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.tight_layout()

# Show plot
plt.show()
```
